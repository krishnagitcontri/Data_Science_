## Distribution Plot

- Distplot - merges density and histogram
- Joinplot
- KDEplot
- Pairplot
- Rugplot
- Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Distplot
- Distribution of a univariant value (single value)

```python
sns.distplot(tips['total_bill'], kde = True)
```

## Jointplot
### Kind: 
- hex
- hist
- kde
- scatter
- reg
- resid

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter')
```

## Pairplot

```python
sns.pairplot(tips, hue = 'sex', palette = 'viridis')
```

## Rugplot

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.heatmap(tips.corr(numeric_only=True), cmap = 'bone', annot = True)
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker', palette='bone')
```

```python
sns.boxplot(data=tips, palette='bone', orient='h')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='bone')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='smoker', palette='bone')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker', palette='bone')
```
