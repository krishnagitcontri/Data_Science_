## Distribution Plot

- **Distplot** - A combination of a histogram and a KDE (Kernel Density Estimate) plot that visualizes the distribution of a *univariate* set of observations.

- **Joinplot** - A plot that presents *bivariate* scatterplots and univariate histograms in the same figure, ideal for comparing two distributions.

- **KDEplot** - A plot that visualizes the probability density of a *continuous variable*, essentially smoothing a histogram.

- **Pairplot** - A plot that displays pairwise relationships across an entire dataframe, useful for *spotting trends and correlations*.

- **Rugplot** - A plot that puts a tick for every data point on a *univariate* distribution, providing a granular view.

- **Heatmap** - A two-dimensional graphical representation of data where the *individual* values are represented as colors, often used for correlation matrices.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Distplot
- Distribution of a univariant value (single value)

```python
sns.distplot(tips['total_bill'], kde = True)
```

## Jointplot
### Kind: 
- hex
- hist
- kde
- scatter
- reg
- resid

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter')
```

## Pairplot

```python
sns.pairplot(tips, hue = 'sex', palette = 'viridis')
```

## Rugplot

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.heatmap(tips.corr(numeric_only=True), cmap = 'bone', annot = True)
```
